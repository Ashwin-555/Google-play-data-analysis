```python
mean_value=data["Rating"].mean()
data["Rating"].fillna(mean_value,inplace=True)
```

```python
#Duplicate value
duplicate_values= data.duplicated().sum()
print(duplicate_values)
```

```python
#value_counts() - count of occurence of each unique values in the particular column
data["Category"].value_counts()
```

```python
#unique{}- take unique values outside the column
data["Category"].unique()
```

```python
data.dropna(subset=["Category"],inplace=True)
data.dropna(subset=["Category"],inplace=True)
data["Category"].isnull().sum()
```

```python
#histogram plot for rating distribution
plt.figure(figsize=(10,15))
sns.histplot(data["Rating"].dropna(),kde=True,bins=20,color="Yellow")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
#box plot for the rating distribution
plt.figure(figsize=(10,6))
plt.title("Box plot for ratings")
sns.boxplot(x="Rating",data=data,color="skyblue")
plt.xlabel("Rating")
plt.show()
```

```python
#to remove outliers we can use two methods
#1.log1p() method - outliers only on right or left of the box plot
#2.IQR method - outliers on both sides of the box plots
```

```python
#using log1p method in box plot
data["Rating"]=np.log1p(data["Rating"])
```

```python
#box plot for the rating distribution
plt.figure(figsize=(10,6))
plt.title("Box plot for ratings")
sns.boxplot(x="Rating",data=data,color="skyblue")
plt.xlabel("Rating")
plt.show()
```

```python
#Using IQR method in box plot
Q1=data["Rating"].quantile(0.25)
Q2=data["Rating"].quantile(0.5)
Q3=data["Rating"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q1+1.5*IQR
```

```python
#to remove ouliers and have filtering the valid data inside the variable
filterdata=data[(data["Rating"]>=lower_bound) & (data["Rating"]<=upper_bound)]
```

```python
#display the outlier free box plot
plt.figure(figsize=(10,6))
plt.title("OUTLIER FREE BOX PLOT")
sns.boxplot(x="Rating",data=filterdata,color="lightgreen")
plt.xlabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y="Category",data=filterdata,order=data["Category"].value_counts().index,color="red")
plt.title("COUNT PLOT of APP CATEGORIES")
plt.xlabel("count")
plt.ylabel("Category")
plt.show()
```

```python
#heatmap for the entire dataset
plt.figure(figsize=(16,6))
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(16,6))
missing_values=data.isnull().sum()/len(data)*100
missing_values.plot(kind='bar')
plt.xlabel('count')
plt.ylabel('percentage')
plt.title("Missing values in the data")
plt.show()
```

```python
#Percentage of missing values in each column
(data.isnull().sum()/len(data)*100).sort_values(ascending=False)
```
